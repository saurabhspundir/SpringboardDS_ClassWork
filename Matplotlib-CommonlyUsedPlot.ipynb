```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# histogram
fig,ax = plt.subplots()

# creating sample
x = np.random.normal(size=1000)

# histogram , x number of bins, type of histogram: bar,step, step filled, barstacked
ax.hist(x,bins=50,alpha=0.5,histtype='stepfilled')
```

```python
# pie plot
fraction =[10,20,35,5,15,10,5]
fig,ax =plt.subplots()

# draw pie plot, frastions is data, labels to amrk. can give colors too
ax.pie(fraction,labels=['food','travel','movie','grocery','rent','car','gas'])
```

```python
# error plots
fig,ax = plt.subplots()
x = np.linspace(0, 10, 30)
dy = 0.1
y = np.random.normal(np.sin(x), dy)

ax.errorbar(x,y,dy,fmt='.k')
```

```python
# filled plot
fig, ax = plt.subplots()
x = np.linspace(0, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

# values(xmy1,y2). condition where to fill, color type. if no where and color. it fills deafault
plt.fill_between(x,y1,y2, where=(y1 < y2),color='red')
plt.fill_between(x,y1,y2, where=(y1 > y2),color='blue')
```

```python
# scatter plot ** used
fig,ax = plt.subplots()
x = np.random.random(50)
y = np.random.random(50)
c = np.random.random(50)  # color of points
s = 500 * np.random.random(50)  # size of points
# scatter vales(y,x or x,y), color array to use, size array to use 
scbar_img=ax.scatter(y,x,c=c,s=s,cmap=plt.cm.jet)

# add color bar to right
fig.colorbar(scbar_img, ax=ax)

# set the color limits , override dfefault
scbar_img.set_clim(0.0, 1.0)
```

```python
# contour plots
x = np.linspace(0, 10, 50)
y = np.linspace(0, 20, 60)

z = np.cos(y[:, np.newaxis]) * np.sin(x)

fig, ax = plt.subplots()

# filled contours. This make shapes
im = ax.contourf(x, y, z, 100)

# contour lines. This fill colors in shape
im2 = ax.contour(x, y, z, colors='k')

fig.colorbar(im, ax=ax)
```

```python
x, y = np.random.normal(size=(2, 10000))

fig, ax = plt.subplots()
im = ax.hexbin(x, y, gridsize=20)
fig.colorbar(im, ax=ax)

fig, ax = plt.subplots()
H = ax.hist2d(x, y, bins=20)
fig.colorbar(H[3], ax=ax)
```

```python
# polar plot

fig = plt.figure()
# create a empty cirle plot 
ax = fig.add_subplot(1, 1, 1, projection='polar')
                    
theta = np.linspace(0, 10 * np.pi, 1000)
r = np.linspace(0, 10, 1000)

ax.plot(theta, r)
```
